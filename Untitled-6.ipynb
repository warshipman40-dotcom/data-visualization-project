```python
#imports a speceific plotting function from the matplotlib library
#imports as plt for convenience 
import matplotlib.pyplot as plt
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
#plots the squares and shows it
#because we provided the input values, plot() generates starting from 1, instead of assuming 0
plt.plot(input_values, squares, linewidth = 5)
#set chart title and labels axes
plt.title("Square Numbers", fontsize = 24)
plt.xlabel("Value", fontsize = 14)
plt.ylabel("Square of Value", fontsize = 14)

#Set size of tick labels
plt.tick_params(axis="both", labelsize = 14)
plt.show()

```
